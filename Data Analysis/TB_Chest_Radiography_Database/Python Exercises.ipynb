## Regular expression to match a line that doesn't contain a word
https://stackoverflow.com/questions/406230/regular-expression-to-match-a-line-that-doesnt-contain-a-word

```python
r'^((?!hede).)*$'
# The regex above will match any string, 
# or line without a line break, not containing the (sub)string 'hede'. 
841

```

```python
#Note that the solution to does not start with “hede”:

r'^(?!hede).*$'
#is generally much more efficient than the solution to does not contain “hede”:

r'^((?!hede).)*$'
# The former checks for “hede” only at the input string’s first position, 
# rather than at every position.
```

```python
import numpy as np
import pandas as pd

normal =pd.read_excel('Normal.metadata.xlsx')
tb =pd.read_excel('Tuberculosis.metadata.xlsx')
```

```python
display(normal[:500])
display(tb[:500])
```

```python
normal_train = normal[:500]
normal_train['class_name'] = 'Normal'
normal_train = normal_train.rename(columns={'FILE NAME':'image_id'})
normal_train
# normal_train.loc[:,'class_name'] = 'Normal'
```

```python
tb_train = tb[:500]
tb_train['class_name'] = 'Tuberculosis'
tb_train = tb_train.rename(columns={'FILE NAME':'image_id'})
tb_train
```
